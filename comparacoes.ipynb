```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import random
import importlib.util
from pathlib import Path

#configuracoes
BASE_DIR = "svd_melspec"
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
EPOCHS = 20
SEED = 42


def set_all_seeds(seed):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)


def collect_generator_arrays(generator):
    generator.reset()
    x_batches, y_batches = [], []

    for idx in range(len(generator)):
        x_batch, y_batch = generator[idx]
        x_batches.append(x_batch)
        y_batches.append(y_batch)

    x_data = np.concatenate(x_batches, axis=0)
    y_data = np.concatenate(y_batches, axis=0).astype(int)
    return x_data, y_data


set_all_seeds(SEED)

#carregando mascaras
def _load_augmentation_masks():
    module_path = Path("config\data_augmentation.py")
    spec = importlib.util.spec_from_file_location("data_augmentation_copy", module_path)
    if spec is None or spec.loader is None:
        raise ImportError(f"Nao foi possivel carregar: {module_path}")
    module = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(module)
    return module.time_mask, module.freq_mask


base_time_mask, base_freq_mask = _load_augmentation_masks()


def time_mask(img_array):
    img_array = img_array.copy()

    if img_array.ndim == 3 and img_array.shape[-1] == 1:
        img_array[:, :, 0] = base_time_mask(img_array[:, :, 0])
        return img_array
    return base_time_mask(img_array)


def freq_mask(img_array):
    img_array = img_array.copy()

    if img_array.ndim == 3 and img_array.shape[-1] == 1:
        img_array[:, :, 0] = base_freq_mask(img_array[:, :, 0])
        return img_array
    return base_freq_mask(img_array)


def time_freq_mask(img_array):
    img_array = time_mask(img_array)
    img_array = freq_mask(img_array)
    return img_array



def make_generators(mask_type):
    if mask_type == "sem_mascara":
        preprocessing_fn = None
    elif mask_type == "temporal":
        preprocessing_fn = time_mask
    elif mask_type == "frequencia":
        preprocessing_fn = freq_mask
    elif mask_type == "temporal_frequencia":
        preprocessing_fn = time_freq_mask
    else:
        raise ValueError(f"mask_type invalido: {mask_type}")

    train_datagen = ImageDataGenerator(
        rescale=1.0 / 255,
        preprocessing_function=preprocessing_fn
    )
    val_datagen = ImageDataGenerator(rescale=1.0 / 255)
    test_datagen = ImageDataGenerator(rescale=1.0 / 255)

    train_generator = train_datagen.flow_from_directory(
        f"{BASE_DIR}/train",
        target_size=IMG_SIZE,
        color_mode="grayscale",
        batch_size=BATCH_SIZE,
        class_mode="binary"
    )

    val_generator = val_datagen.flow_from_directory(
        f"{BASE_DIR}/validation",
        target_size=IMG_SIZE,
        color_mode="grayscale",
        batch_size=BATCH_SIZE,
        class_mode="binary"
    )

    test_generator = test_datagen.flow_from_directory(
        f"{BASE_DIR}/test",
        target_size=IMG_SIZE,
        color_mode="grayscale",
        batch_size=BATCH_SIZE,
        class_mode="binary",
        shuffle=False
    )

    class_labels = [None] * len(test_generator.class_indices)
    for class_name, class_idx in test_generator.class_indices.items():
        class_labels[class_idx] = class_name

    return train_generator, val_generator, test_generator, class_labels


#construindo rede neural
def build_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation="relu", input_shape=(128, 128, 1)),
        MaxPooling2D(2, 2),

        Conv2D(64, (3, 3), activation="relu"),
        MaxPooling2D(2, 2),

        Flatten(),

        Dense(64, activation="relu"),
        Dropout(0.5),

        Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model


#treino + teste
def run_experiment(mask_type):
    if mask_type == "temporal_frequencia":
        display_name = "temporal + frequencia"
    else:
        display_name = mask_type.replace("_", " ")
    print(f"\n===== EXPERIMENTO: {display_name} =====")

    tf.keras.backend.clear_session()
    set_all_seeds(SEED)

    train_generator, val_generator, test_generator, class_labels = make_generators(mask_type)

    early_stop = EarlyStopping(
        monitor="val_loss",
        patience=10,
        restore_best_weights=True
    )

    model = build_model()
    model.summary()

    history = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=EPOCHS,
        callbacks=[early_stop]
    )

    x_test, y_true = collect_generator_arrays(test_generator)
    loss, acc = model.evaluate(x_test, y_true, batch_size=BATCH_SIZE, verbose=0)

    y_pred_prob = model(x_test, training=False).numpy().ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)

    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    print(f"Acuracia no teste ({display_name}): {acc:.4f}")
    print("\nRelatorio de Classificacao:")
    print(classification_report(y_true, y_pred, target_names=class_labels, zero_division=0))

    cm = confusion_matrix(y_true, y_pred)

    return {
        "mascara": display_name,
        "loss": loss,
        "accuracy": acc,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "cm": cm,
        "history": history,
        "class_labels": class_labels,
    }


#teste de metodos
experiments = []
for mask in ["sem_mascara", "temporal", "frequencia", "temporal_frequencia"]:
    experiments.append(run_experiment(mask))


#comparacao de mtricas
metrics_df = pd.DataFrame([
    {
        "mascara": exp["mascara"],
        "loss": exp["loss"],
        "accuracy": exp["accuracy"],
        "precision": exp["precision"],
        "recall": exp["recall"],
        "f1": exp["f1"],
    }
    for exp in experiments
])

metrics_df = metrics_df.sort_values(by="f1", ascending=False).reset_index(drop=True)
print("\nComparativo final (ordenado por F1):")
display(metrics_df)


#matrizes de confusão
fig, axes = plt.subplots(1, len(experiments), figsize=(5 * len(experiments), 4))
if len(experiments) == 1:
    axes = [axes]

for idx, exp in enumerate(experiments):
    sns.heatmap(
        exp["cm"],
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=exp["class_labels"],
        yticklabels=exp["class_labels"],
        ax=axes[idx],
        cbar=False,
    )
    axes[idx].set_title(f"Matriz de Confusao - {exp['mascara']}")
    axes[idx].set_xlabel("Predito")
    axes[idx].set_ylabel("Real")

plt.tight_layout()
plt.show()



plt.figure(figsize=(8, 5))
for exp in experiments:
    plt.plot(exp["history"].history["val_accuracy"], label=f"val - {exp['mascara']}")

plt.title("Comparacao de val_accuracy por mascara")
plt.xlabel("Epoca")
plt.ylabel("Acuracia")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(8, 5))
for exp in experiments:
    plt.plot(exp["history"].history["val_loss"], label=f"val_loss - {exp['mascara']}")

plt.title("Comparacao de val_loss por mascara")
plt.xlabel("Epoca")
plt.ylabel("Loss")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# 1) barras comparativas das metricas principais
plot_df = metrics_df.melt(
    id_vars="mascara",
    value_vars=["accuracy", "precision", "recall", "f1"],
    var_name="metrica",
    value_name="valor",
)

plt.figure(figsize=(11, 5))
ax = sns.barplot(data=plot_df, x="mascara", y="valor", hue="metrica")
for patch in ax.patches:
    height = patch.get_height()
    if np.isfinite(height):
        ax.annotate(
            f"{height:.3f}",
            (patch.get_x() + patch.get_width() / 2, height),
            ha="center",
            va="bottom",
            fontsize=8,
            xytext=(0, 3),
            textcoords="offset points",
        )
plt.title("Comparacao de metricas principais por mascara")
plt.xlabel("Mascara")
plt.ylabel("Valor")
plt.ylim(0, 1)
plt.grid(axis="y", alpha=0.25)
plt.legend(title="Metrica")
plt.tight_layout()
plt.show()

# 2) bump chart de ranking por metrica principal (1 = melhor)
rank_metrics = ["accuracy", "precision", "recall", "f1"]
rank_df = metrics_df.set_index("mascara")[rank_metrics].rank(method="min", ascending=False).T

fig, ax = plt.subplots(figsize=(10, 6))
palette = sns.color_palette("Set2", n_colors=len(rank_df.columns))

for mascara, color in zip(rank_df.columns, palette):
    y_vals = rank_df[mascara].values
    ax.plot(rank_metrics, y_vals, marker="o", linewidth=2.5, markersize=8, label=mascara, color=color)
    for x_metric, y_rank in zip(rank_metrics, y_vals):
        ax.text(x_metric, y_rank - 0.08, f"{int(y_rank)}", ha="center", va="top", fontsize=8, color=color)

ax.set_title("Bump Chart: Ranking por metrica principal")
ax.set_xlabel("Metrica")
ax.set_ylabel("Posicao no ranking")
ax.set_yticks(range(1, len(rank_df.columns) + 1))
ax.set_ylim(len(rank_df.columns) + 0.3, 0.7)
ax.grid(axis="y", alpha=0.25, linestyle="--")
ax.legend(title="Mascara", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()

# 3) curva de validacao (accuracy e loss)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for exp in experiments:
    nome = exp["mascara"]
    val_acc = exp["history"].history.get("val_accuracy", [])
    val_loss = exp["history"].history.get("val_loss", [])

    axes[0].plot(val_acc, linewidth=2, label=f"val - {nome}")
    axes[1].plot(val_loss, linewidth=2, label=f"val - {nome}")

axes[0].set_title("Curvas de validacao - accuracy")
axes[0].set_xlabel("Epoca")
axes[0].set_ylabel("Accuracy")
axes[0].grid(alpha=0.25)

axes[1].set_title("Curvas de validacao - loss")
axes[1].set_xlabel("Epoca")
axes[1].set_ylabel("Loss")
axes[1].grid(alpha=0.25)

handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", ncol=4, bbox_to_anchor=(0.5, 1.08))
plt.tight_layout()
plt.show()
```
